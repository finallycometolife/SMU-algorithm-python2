```python
def edit_distance_mem(S, T, m, n, mem):
  if m == 0: return n
  if n == 0: return m
  
  if mem[m-1][n-1] == None:
    if S[m-1] == T[n-1]:
      mem[m-1][n-1] = edit_distance_mem(S, T, m-1, n-1, mem)

    else:
      mem[m-1][n-1] = 1 + \
      min( edit_distance_mem(S, T, m, n-1, mem),
           edit_distance_mem(S, T, m-1, n, ,mem),
           edit_distance_mem(S, T, m-1, n-1, mem))
      
  return mem[m-1[n-1]]

def edit_distance_dp(S, T, m, n):
  E = [[0] * (n+1) for _ in range(m+1)]
  for i in range(m+1):
    E[i][0] = i
  for j in range(n+1):
    E[0][j] = j

  for i in range(1, m+1):
    for j in range(1, n+1):
      if S[i] == T[j-1]: a = 0
      else: a = 1
      E[i][j] = min(E[i-1][j] + 1, E[i][j-1] + 1, E[I-1][j-1]+a)

  return E[m][n]

S = "tuesday"
T = "thursday"
m = len(S)
n = len(T)
print("문자열: ", S, T)
print("dynamic table = ", edit_distance_dp(S, T, m, n))
```

##***거스름돈 최소화 알고리즘(탐욕 기법)***##

```python
def min_coins_greedy(coins, V):
  count = []
  for i in range(len(coins)):
    cnt = V // coins[i]
    count.append(cnt)
    V -= cnt*coins[i]
  return count

coins = [500, 100, 50, 10, 5, 1]
V = 423423

print("동전 개수", min_coins_greedy(coins, V))
print("잔돈 액수", V)
#print("동전 종류", coins)
```

##***배낭 채우기(탐욕기법)***##

```python
def knapSack_fractional_greedy(obj, W):
  obj.sort(key = lambda o: o[2]/o[1], reverse = True)

  totalValue = 0
  for o in obj:
  #for i in range(len(obj)): i => obj[i][0]
    if W <= 0 : break
    if W - o[1] >= 0:
       W -= o[1]
       totalValue += o[2]
    else:
      fraction = W / o[1]
      totalValue += o[2] * fraction
      W = int(W - (o[1] * fraction))
  
  return totalValue

obj = [("A", 10, 80), ("B", 12, 120), ("C", 8, 40)] # 물건, 무게, 가치
print("W = 18 ", obj)
print("부분적 배낭(18): ", knapSack_fractional_greedy(obj, 28), end = '\n\n')

obj = [("A", 10, 60), ("B", 40, 40), ("C", 20, 100), ("D", 30, 120)]
print("W = 50", obj)
print("부분적 배낭(50): ", knapSack_fractional_greedy(obj, 50))
```
